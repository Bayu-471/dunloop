```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import cv2

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
```

```python
citra1 = imread(fname="noisyimage-copy1.jpg")
print(citra1.shape)
plt.imshow(citra1,cmap="gray")
```

```python
kernel = np.array([[-1,0,-1],[0,4,0],[-1,0,-1]])
citraoutput = cv2.filter2D(citra1,-1,kernel)
fig,axes = plt.subplots(1,2,figsize=(15,11))
ax = axes.ravel()

ax[0].imshow(citra1,cmap="gray")
ax[0].set_title("citra awal")
ax[1].imshow(citraoutput,cmap="gray")
ax[1].set_title("citra awal")
```
